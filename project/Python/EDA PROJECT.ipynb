```python
df1 = df.groupby('Category')['Quantity'].sum().reset_index()
df1
```

```python
# matplotlib
plt.figure(figsize=(10,5))
plt.plot(df1['Category'],df1['Quantity'],marker="*",color='blue',linestyle='--')
plt.title("Cut wise price")
plt.xlabel("cut")
plt.grid(True)
plt.ylabel("price")
plt.show()
```

```python
# seaborn
sns.lineplot(data=df1, x=df1['Category'],y=df1['Quantity'], marker='o', color='red')
plt.grid(True)
plt.show()
```
